```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
amazon = pd.read_csv(r"C:\Users\adity\Downloads\amazon_jobs_dataset.csv")
```

```python
amz = amazon.copy()
amz.head()
```

```python
amz.info()
```

```python
amz['Year'] = pd.to_datetime(amz['Posting_date']).dt.year 
# Used pd.to_datetime function to convert Posting date into numerical values, then extracted the Year from Posting_date using .dt.year attribute.
```

```python
amz
```

```python
# A1. 

total_years = amz.Year.value_counts().sort_index().index 
# Filtered Years, then counted each one, sorted to show in ascending, and only took the index.

year_wise_job_posting = amz.Year.value_counts().sort_index().values
# Same as above, but this time took values.

plt.plot(total_years, year_wise_job_posting)
# To plot a line graph, use plt.plot(X-axis, Y-axis).

plt.xlabel("Years") # For Showing X-Label.
plt.ylabel("No. of Job Postings") # For Showing Y-Label.
plt.title("Job-Year Plot") # For the Title of the Graph.
plt.show() # Hides Unneccessary information.
```

```python
amz['Month'] = pd.to_datetime(amz['Posting_date']).dt.month_name() # Used the same method as Year to extract Month.
amz 
```

```python
# A2.

months = amz.Month.value_counts().index # Counted the months, stored the index.

month_wise_job_postings = amz.Month.value_counts().values # Stored its values in this one.

plt.bar(months, month_wise_job_postings) # PLotted the Bar Graph. 

plt.xlabel("Months") # For Labeling the X-axis.
plt.ylabel("No. of Job Postings") # For Labeling the Y-axis.
plt.title("Job-Month Plot") # The Title of the Graph.
plt.xticks(rotation = 45) # This function rotates the X labels to the specified degree.
plt.show() # For only showing the Plot.
```

```python
# For Showing it in Calender Order.

calender_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# Specify the Order by making a list of it.

months_by_calender_order = amz.Month.value_counts().reindex(calender_order).index
# Use .reindex() function to adjust the index of the values. Then, stored the index of the result.

job_postings_calender_order = amz.Month.value_counts().reindex(calender_order).values
# Stored its values in another variable.

plt.bar(months_by_calender_order, job_postings_calender_order)
# Plotted the Bar Groph.

plt.xticks(rotation = 45)
# Used xticks() function to adjust the X labels.

plt.xlabel("Months By Calender Order")   # For Labeling the X-axis.
plt.ylabel("No. of Job Postings")        # For Labeling the Y-axis.
plt.title("Job-Month Plot")              # The Title of the Graph.

plt.show()
```

```python
amz['City'] = amz.location.str.split(",").str[2]  
# To extract the City to use in the question, splitted the location by "," and extracted the second index, i.e., City.
```

```python
amz
```

```python
# A3. 

city = amz['City'][amz.location.str.startswith("IN")].value_counts().index
# Filtered to show City in India, then counted them, then stored its index

city_wise_job_posting = amz['City'][amz.location.str.startswith("IN")].value_counts().values
# Stored the values here.

plt.pie(city_wise_job_posting, labels = city, autopct = "%.2f%%")
# Plotting the Pie Chart. Labelled according to the City, autopct attribute for showing percentage upto 2 decimal points.
plt.show()
```

```python
total_jobs = amz['City'][amz.location.str.startswith("IN")].value_counts().values.sum()
# For calculating the percentages later.

percentages = (city_wise_job_posting*100 / total_jobs).round(2)
# Calculating the percentages.

for i in range(len(city)):
    print(city[i], percentages[i])
```

```python
# A4.

year_wise_java_job = amz.Year[amz['BASIC QUALIFICATIONS'].str.contains("Java", case = False, na = False)].value_counts().sort_index()
# Filtered Year Wise and searched for Java in Basic Qualifications where case-sensitivity is False and null values are not counted.
# Then counted them and sorted for future use in Graphs.

job_postings_in_java = year_wise_java_job.index
# Stored the index to use as X-axis.

year = year_wise_java_job.values
# Stored the values to use as Y-axis.

plt.scatter(job_postings_in_java, year)
# Plotted the Scatter Graph.

plt.show()
```

```python
# Loop for showing it in numbers.
for i in range(len(job_postings_in_java)):
    print(job_postings_in_java[i], year[i])
```
