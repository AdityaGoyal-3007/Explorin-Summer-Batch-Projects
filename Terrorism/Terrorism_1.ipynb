```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
terrorism = pd.read_csv(r"terrorismData.csv")
terrorism.head()
```

```python
tr = terrorism.copy()
```

```python
# A1.
MDC = tr.City[tr.State == "Jammu and Kashmir"].value_counts().index[0]

MAGIJK = tr.Group[(tr.State == "Jammu and Kashmir") & (tr.Group != "Unknown") & (tr.City == MDC)].value_counts().index[0]

NoA = tr.City[tr.State == "Jammu and Kashmir"].value_counts().values[0]

print(MDC,"|", NoA,"|", MAGIJK)
```

```python
# A2. 
Highest_Terror_Attack_Country = tr.Country.value_counts().index[0]

Most_Attacked_Year = tr.Year[tr.Country == Highest_Terror_Attack_Country].value_counts().index[0]

NoA = tr.Country.value_counts().values[0]

print(Highest_Terror_Attack_Country, "|", NoA, "|", Most_Attacked_Year)
```

```python
# A3.
Deadliest_Attack = tr.Killed.max()

Country = tr.Country[tr.Killed == Deadliest_Attack].values[0]

Terrorist_Group = tr.Group[(tr.Country == Country) & (tr.Killed == Deadliest_Attack)].values[0]

print(int(Deadliest_Attack), "|", Country, "|", Terrorist_Group)
```

```python
# Made a new column in the DataFrame to work with Dates.
real_months = tr.Month.replace(0, 1)
real_days = tr.Day.replace(0,1)

tr['Date'] = pd.to_datetime(dict(year = tr.Year, month = real_months, day = real_days))
tr
```

```python
# A4.
Time_period_start = pd.Timestamp(2014, 5, 26)

Total_NoA_After = len(tr[(tr.Country == "India") & (tr.Date >= Time_period_start)]
                     )
Most_Active_Group_in_Time_period = tr.Group[(tr.Country == "India") & (tr.Date >= Time_period_start) & (tr.Group != "Unknown")].value_counts().index[0]

print(Total_NoA_After, Most_Active_Group_in_Time_period)
```

```python
Total_Casuality_In_Red_Corridor_States = tr.Casuality[(tr.State == "Jharkhand") | (tr.State == "Odisha") | (tr.State == "Andhra Pradesh") | (tr.State == "Orissa") | (tr.State == "Andhra pradesh") | (tr.State == "Chhattisgarh")].sum()

terror1 = tr[(tr.State == "Jharkhand") | (tr.State == "Odisha") | (tr.State == "Andhra Pradesh") | (tr.State == "Orissa") | (tr.State == "Andhra pradesh") | (tr.State == "Chhattisgarh")]

Total_Years = max(terror1.Year) - min(terror1.Year) + 1

Frequency_For_Red_Corridor_States = int(Total_Casuality_In_Red_Corridor_States/Total_Years)

TCJK = tr.Casuality[tr.State == "Jammu and Kashmir"].sum()

terror2 = tr[tr.State == "Jammu and Kashmir"]

Total_Years2 = max(terror2.Year) - min(terror2.Year) + 1

Frequency_For_Jammu_and_Kashmir = int(TCJK/Total_Years2)

print(Frequency_For_Red_Corridor_States, Frequency_For_Jammu_and_Kashmir)
```
