```python
import numpy as np
import pandas as pd
```

```python
terrorism = pd.read_csv(r"C:\Users\adity\Downloads\terrorismData.csv")
terrorism.head()
ter = terrorism.copy()
```

```python
# Cleaning

del terrorism['Latitude']
del terrorism['Longitude']
del terrorism["AttackType"]
del terrorism['Target']
del terrorism['Summary']
del terrorism['Target_type']
del terrorism['Weapon_type']

terrorism.fillna({'Killed' : 0}, inplace = True)
terrorism.fillna({'Wounded' : 0}, inplace = True)
```

```python
# 1. Find the most frequent day of attack in terrorismDataset. 
# Print count of frequent day and number of attacks as integer values. 

Day = terrorism['Day'].value_counts().index[0]
count = terrorism['Day'].value_counts().values[0]
print(Day, count)
```

```python
# 2. Find number of attacks between 1 Jan 2010 and 31 Jan 2010 (both inclusive).
# Note: Ignore cases where day = 0.

NoAiJ = terrorism.Year[(terrorism.Day >= 1) & (terrorism.Month == 1) & (terrorism.Year == 2010)].value_counts().values[0]
# NoAiJ = len(terrorism[(terrorism.Day >= 1) & (terrorism.Month == 1) & (terrorism.Year == 2010)])
print(NoAiJ)
```

```python
# 3. Find the attack between May 1999 and July 1999 (Kargil War Period which happened in Jammu and Kashmir) with maximum casualties.
# Casualty = Killed + Wounded (fill missing values with 0).

terrorism['Casuality'] = terrorism['Killed'] + terrorism['Wounded']
jk_57 = terrorism[(terrorism.State == "Jammu and Kashmir") & (terrorism.Year == 1999) & (terrorism.Month.isin([5,6,7]))]
max_kill = jk_57.loc[jk_57.Casuality.idxmax()]
print(max_kill['Casuality'], max_kill['City'], max_kill['Group'])
```

```python
# 4. Find top 5 Indian cities with most casualties.
# Ignore unknown cities.

indian_df = terrorism[(terrorism.Country == "India") & (terrorism.City != "Unknown")].copy()
top_5_cities = indian_df.groupby("City")['Casuality'].sum().sort_values(ascending=False).head(5)
print(top_5_cities)
```
